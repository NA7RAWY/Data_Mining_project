```python
import pandas as pd

#Load the dataset
file_path = "E:/faculty/3 first term/kaggle/Life Expectancy Data.csv"
df = pd.read_csv(file_path)

#Display the first few rows of the dataset to understand its structure
df.head()
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df.head()
```

```python
#Drop irrelevant columns (e.g., 'Country')
df_clean = df.drop(columns=['Country'])

#Encode the 'Status' column (convert 'Developing' and 'Developed' to numerical values)
le = LabelEncoder()
df_clean['Status'] = le.fit_transform(df_clean['Status'])

#Handling missing values by filling them with the mean (could also drop or use other strategies)
df_clean.fillna(df_clean.mean(), inplace=True)

#Separate features and target variable (assuming 'Life expectancy' is the target)
X = df_clean.drop(columns=['Life expectancy '])
y = df_clean['Life expectancy ']

#Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Convert the standardized features back into a DataFrame for clarity
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

#Display the processed features and target variable
X_scaled_df.head(), y.head()
```

```python
# Combine the scaled features and target variable into one DataFrame
preprocessed_df = pd.concat([X_scaled_df, y.reset_index(drop=True)], axis=1)

# Save the preprocessed DataFrame to a CSV file
output_file_path = "E:/faculty/3 first term/kaggle/Preprocessed_Life_Expectancy_Preprocessed.csv"
preprocessed_df.to_csv(output_file_path, index=False)
```
