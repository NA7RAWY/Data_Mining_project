```python
import pandas as pd
```

```python
df=pd.read_csv("gym_members_data_with_missing_values.csv")
```

```python
df.head(10)
```

```python
null_count = df.isnull().sum().sum()

print(null_count)
```

```python
numerical_features = ['Weight (kg)', 'Height (m)', 'Session_Duration (hours)']
categorical_features = ['Gender', 'Workout_Type']
```

```python
for col in numerical_features:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
for col in categorical_features:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df = pd.get_dummies(df, columns=categorical_features, drop_first=True)
```

```python
for col in numerical_features:
    df[col] = (df[col] - df[col].mean()) / df[col].std()
```

```python
print(df.head())
```

```python
# Save the processed data to a new CSV file
output_file_path = 'processed_gym_members_data.csv'
df.to_csv(output_file_path, index=False)

output_file_path
```
