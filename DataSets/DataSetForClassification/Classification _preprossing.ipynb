```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, OneHotEncoder
```

```python
# Step 1: Normalize numerical features
scaler = MinMaxScaler()
numerical_cols = ['Study Hours per Week', 'Attendance Rate', 'Previous Grades']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
# Step 2: Encode binary categorical feature 'Participation in Extracurricular Activities'
df['Participation in Extracurricular Activities'] = df['Participation in Extracurricular Activities'].astype(int)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Step 3: OneHotEncode 'Parent Education Level'
encoder = OneHotEncoder(sparse_output=False)
encoded_cols = encoder.fit_transform(df[['Parent Education Level']])

# Get the new column names for the one-hot encoded features
encoded_col_names = encoder.get_feature_names_out(['Parent Education Level'])

# Convert the one-hot encoded array into a DataFrame
encoded_df = pd.DataFrame(encoded_cols, columns=encoded_col_names)

# Concatenate the new encoded columns back to the original DataFrame
df = pd.concat([df, encoded_df], axis=1).drop(columns=['Parent Education Level'])

```

```python
# Step 4: Encode target column 'Passed'
label_encoder = LabelEncoder()
df['Passed'] = label_encoder.fit_transform(df['Passed'])
```

```python
df.to_csv("output/cleaned-student_performance_prediction.csv", index=False)
```
